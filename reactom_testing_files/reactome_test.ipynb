```python
import requests
import pandas as pd
import json
```

```python
url = "https://reactome.org/ContentService/"

xrefQuery = "references/mapping/{queryID}"
queryQuery = "data/query/{reactomeID}/physicalEntity"
pathwayQuery = "data/pathways/low/entity/{reactomeID}?species={species}"
```

For Each entry one needs two queries (which isnt great)

```python
#original Data
testUniprot = ["Q8BWN8",
"P47740",
"Q9R0H0",
"Q9DBK0",
"O88844",
"Q9D379",
"P34914",
"Q9QZD8",
"O08756",
"P45952"
]
```

```python
#Urls for xref (getting the reactom DB ID)
xrefUrls = [url + xrefQuery.format(queryID=ID) for ID in testUniprot]
xrefUrls 
```

```python
#get xrefRequest (VERY VERY SLOW)
xrefs = [requests.get(url).json() for url in xrefUrls]
xrefs
```

```python
#extract interal reactome ids (numeric)
internalIDs = {elem['identifier']: elem['dbId'] for elemOuter in xrefs for elem in elemOuter }
internalIDs
```

```python
#build query Querys (getting the Reactome Ids (e.g. R-ABC-123456))
queryUrls = {k: url + queryQuery.format(reactomeID=v) for k,v in internalIDs.items()}
queryUrls
```

```python
#request texts 
requests_texts = {k:requests.get(v).text for k,v in queryUrls.items()}
requests_texts
```

```python
#uniprot IDs to lists containing dicts with {'reactomeID': string, 'symbolLoc': string, 'type': string}
uniprot_to_req_dict = {k: [{'reactomeID': elemOuter.split('\t')[0], 'symbolLoc': elemOuter.split('\t')[1], 'type': elemOuter.split('\t')[2] } for elemOuter in v.strip().split('\n')] for k,v in requests_texts.items()}
uniprot_to_req_dict
```

```python
reactomeID_to_uniprotID = {elem['reactomeID']: k for k, v in uniprot_to_req_dict.items() for elem in v}
reactomeID_to_uniprotID
```

```python
pathway_urls = {k: url + pathwayQuery.format(reactomeID = k, species="mus%20musculus") for k,v in reactomeID_to_uniprotID.items()}
pathway_urls
```

```python
reactomeID_to_reactome_pathways = {k: requests.get(v).json() for k,v in pathway_urls.items() }
```

```python
reactomeID_to_reactome_pathways
```

```python
countPathways = 0
pathways = []

for k,v in reactomeID_to_reactome_pathways.items():
    if isinstance(v, list):
        countPathways += len(v)
        for elem in v:
            pathways.append(elem['stId'])
    else:
        print("Entry: {}, no Pathway Found".format(k))

countPathways
```
